# opening a csv file from an URL

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

# sep parameter

```python
# csv = comma separated values
# tsv = tab separated values

# csv file gula comma diye separated thake, tai read_csv() diye ederke read kora jay.
# But tsv file gula tab diye separated thake, ejonno ederke csv diye sorasori read kora jay na.
# sep parameter use korte hoy.


pd.read_csv('movie_titles_metadata.tsv',sep='\t',names=['sno','name','release_year','rating','votes','genres'])

# kintu ekhane ekta somossa holo table e coloumn gular original name ashe ni. 
# first row ta column gular name hoye geche.
# column er name deyar jonno amra names=[] er ekta list use korte pari, jekhane subidha moto column er name dite parbo.
```

# skip bad lines

```python
pd.read_csv('BX-Books.csv',sep=';',encoding='latin-1') # eta ekta semicolon ; separated file . and encoding latin-1 e deya ache. 

# eta run korar por amra ekta line dekhte pai. " Error tokenizing data. C error: Expected 8 fields in line 6451, saw 9"
# that means amader 6451 e 8 ta field thakar kotha sekahe ache 9 ta. tarmane eta ekta bad line. 
# ei somossa dur korte amra on_bad_lines='skip' kore dibo.
```

```python
pd.read_csv('BX-Books.csv', sep=';', encoding="latin-1",on_bad_lines='skip')
```

# dtypes parameter

# handling dates

```python
# kichu dataset e date column er datatype object deya thake. mane string. but amra sei type k date e convert korte pari.
# date e convert korle amra ei datatype er upor onek operation perform korte parbo jeta object er upor parbo na.

pd.read_csv('matches.csv').info()

# ekhane date er datatype object deya. amra eta date banate chai.
```

```python
pd.read_csv('matches.csv',parse_dates=['date']).info() # ekhon 'datetime64[ns]' datatype hoyeche. etar upor onek operation kora jabe. 
```

# Converters

```python
# amra jodi kono coloumn er value replace korte chai onno ekta value diye tahole amra converters use korbo.
# er jonno amader k python er function use korte hobe then converter use korte hobe.    

def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
    
```

```python
pd.read_csv('matches.csv',converters={'team1':rename,'team2':rename})
# amra team1 and team2 column e jotogula Royal Challengers Bangalore name ache sob RCB kore diyechi.
```

# na_values parameter

```python
# er kaj holo amra jodi kothao dash(-) type er kono value pai jetake amra empty hisebe dhorte chai tokhon amra ei parameter use korbo.
# amra dash badeo onno value keo empty or null or NaN banay nite pari.

pd.read_csv('aug_train.csv',na_values=['Male']) # Male diye joto value ache sob amra NaN kore diyechi.
```

# Loading a huge dataset in chunks

```python
# amader kache jodi onek boro ekta dataset thake jetar jonno kina amader system e jayga nai
# tahole amra chunks use kore dataset e koyekta group e vag kore nite pari. and sei group gula te operation korte pari.

dfs = pd.read_csv('aug_train.csv',chunksize=5000)
```

```python
# ekhon dataset er upor ja e korte hobe sob loop diye korte hobe.

for chunk in dfs:
    print(chunk.shape)
```

````python
from sklearn.datasets import make_regression
import numpy as np
````

````python
X,y = make_regression(n_samples = 4, n_features = 1, n_informative = 1, n_targets = 1, noise = 80, random_state = 13)
````

````python
import matplotlib.pyplot as plt
plt.scatter(X,y)

plt.show()
````

- ```Lets apply OLS ðŸ‘‡```

````python
from sklearn.linear_model import LinearRegression
````

````python
lr = LinearRegression()
lr.fit(X,y)
````

````python
lr.coef_
````

````python
lr.intercept_
````

````python
plt.scatter(X,y)
plt.plot(X,lr.predict(X), color = 'red')


plt.show()
````

````python
# Lets apply Gradient Descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b = 0

y_pred = ((78.35 * X) + 100)
````

````python
plt.scatter(X,y)
plt.plot(X,lr.predict(X), color = 'red', label = "OLS")
plt.plot(X,y_pred, color = 'green', label = 'b = 0')



plt.legend()
plt.show()
````

- ekhane red line ta holo train korar por jeta paichi seta, mane desired line.
- green line ta holo current b er value er jonno j line ta seta.
- idea is to match the red line from the green line. 

````python
# dhore nilam m , b er value amra jani . 

m = 78.35
b = 0

loss_slope = -2 * np.sum( y - m * X.ravel() - b)
loss_slope
````

````python
# lets take learning_rate = 0.1

LR = 0.1
step_size = LR * loss_slope
step_size 
````

````python
# Calculating the new intercept

b = b - step_size
b
````

````python
y_pred1 = ((m * X) + b).reshape(4)
````

````python
y_pred1
````

````python
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
````

- ```eta amader first iteration holo. ebar arekta iteration korbo. ```

````python
# iteration - 2

loss_slope = - 2 * np.sum(y-m * X.ravel() - b)
loss_slope 
````

````python
step_size = loss_slope * LR
step_size 
````

````python
b = b - step_size
b
````

````python
y_pred2 = ((m * X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label = 'OLS')
plt.plot(X,y_pred2,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
````

````python
# Iteration 3
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
````

````python
step_size = loss_slope* LR
step_size
````

````python
b = b - step_size
b
````

````python
y_pred3 = ((78.35 * X) + b).reshape(4)

plt.figure(figsize=(15,15))
plt.scatter(X,y)
plt.plot(X,lr.predict(X),color='red',label='OLS')
plt.plot(X,y_pred3,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred2,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
````

- ```ekhane dekha jacche j amader green line ta correction hote hote desired red line er upor uthe jacche, jeta amader main goal chilo .``` 

- ```ei manual iteration gula amra loop use kore korbo. ```

````python
b = -100
m = 78.35
LR = 0.01

epochs = 100

for i in range(epochs):
    loss_slope = -2 * np.sum(y - m * X.ravel() - b)
    step_size = LR * loss_slope
    b = b - step_size


    y_pred = m * X + b
    plt.plot(X,y_pred)

plt.scatter(X,y)
plt.show()
````

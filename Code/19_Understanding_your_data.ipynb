# 4. Are there any missing values.

- To find out how many missing values in our dataset we simply use ```df.isnull().sum()```.
   - This show us the number of missing values of each column in a pandas series form.
   - So we can have an idea whether to replace the null values with any values or just drop the null values.


````python
df.isnull().sum()
````

# 7. How is the correlation between cols.

- Correlation means if there any data or column have effects on another data or column. That means for increasing of any data, if another data increases then it is positive correlation and so for negative correlation. Correlation between same data is 1.
   - If the dataset have only numerical values then you are good to go. Just use ```df.corr()``` and see the results. But if it has both string and numerical values then write ```df.corr(numerical_only=True)``` to select just the numerical values. Otherwise it will show an error.

````python
df.corr(numeric_only=True)
````

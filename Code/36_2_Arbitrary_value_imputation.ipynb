```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
df = pd.read_csv('titanic_toy.csv')
```

```python
df.head()
```

```python
df.isnull().mean()
```

```python
X = df.drop(columns=['Survived'])
y = df['Survived']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
X_train['Age_99'] = X_train['Age'].fillna(99)
X_train['Age_minus1'] = X_train['Age'].fillna(-1)

X_train['Fare_999'] = X_train['Fare'].fillna(999)
X_train['Fare_minus1'] = X_train['Fare'].fillna(-1)


# Missing values gula 99 and -1 diye fill kore dilam.
```

```python
print('Original Age variable variance: ', X_train['Age'].var())
print('Age Variance after 99 wala imputation: ', X_train['Age_99'].var())
print('Age Variance after -1 wala imputation: ', X_train['Age_minus1'].var())

print('Original Fare variable variance: ', X_train['Fare'].var())
print('Fare Variance after 999 wala imputation: ', X_train['Fare_999'].var())
print('Fare Variance after -1 wala imputation: ', X_train['Fare_minus1'].var())
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2,figsize=(12,6))

# For Age.

X_train['Age'].plot(kind='kde',ax=ax[0])
X_train['Age_99'].plot(kind='kde',ax=ax[0])
X_train['Age_minus1'].plot(kind='kde',ax=ax[0])
ax[0].set_title("Age, Age_minus1, Age_99 KDE Plot")
ax[0].legend()

# For Fare

X_train['Fare'].plot(kind='kde',ax=ax[1])
X_train['Fare_999'].plot(kind='kde',ax=ax[1])
X_train['Fare_minus1'].plot(kind='kde',ax=ax[1])
ax[1].set_title("Fare, Fare_minus1, Fare_999 KDE Plot")
ax[1].legend()

plt.show()
```

## Using Sklearn

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
trf = ColumnTransformer(
    [
        ('imputer1',SimpleImputer(strategy = 'constant', fill_value = 99),[0]),
        ('imputer2',SimpleImputer(strategy = 'constant', fill_value = 999),[1])
    ],
    remainder = 'passthrough'
)
```

```python
trf.fit(X_train)
```

```python
trf.named_transformers_['imputer1'].statistics_
```

```python
trf.named_transformers_['imputer2'].statistics_
```

```python
X_train = trf.transform(X_train)
X_test = trf.transform(X_test)
```

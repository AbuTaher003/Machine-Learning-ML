# 3. One Hot Encoding using Sklearn

- ```Pandas use kore one hot encoding korle setay kon column kon position e chilo eta pandas mone rakhe na, tai abar code ta run korle output alada hote pare.```
- ```But Sklearn eta mone rakhe. Tai Machine Learning project korar somoy amra SKlearn use kore One Hot Encoding korbo.```

````python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.iloc[:,0:4],df.iloc[:,-1],
                                                   test_size=0.2,
                                                   random_state=0)

# ekhane first 4 ta column input e niyechi and last column output niyechi.
````

````python
X_train.head()
````

````python
from sklearn.preprocessing import OneHotEncoder
````

````python
ohe = OneHotEncoder(drop='first',sparse_output = False,dtype=np.int32) 

# multicollinearity dur korar jonno amra first column drop kore dilam
# sparse_output = False diye amra sparse matrix k numpy array te convert kore nilam. eta dile ar .toarray() use korar dorkar nai
# dtype=np.int32 ekhon numopy array er sob int hoye jabe
````

````python
X_train_new = ohe.fit_transform(X_train[['fuel','owner']])

# ekhane amra fit and transform eksathe korechi.
# .toarray() korte hobe karon eta initially sparse matrix e thake , tai amra eta numpy array te convert korbo.
````

- ```amra ekhane onehotencoder use kori fuel and owner column er upor , brand and km_driven er upor kori na, tai amra fuel and owner niye alada vabe kaj kori.```
- ```kintu eksomoy sob gula column eksathe jure dite hoy.```
- ```ei bepar ta ektu problematic, eta thik korar upay holo column transfomer, jeta amra porer lecture e dekhbo.```

````python
X_test_new = ohe.transform(X_test[['fuel','owner']])
````

# 4. One Hot Encoding with top Categories

````python
counts = df['brand'].value_counts()
````

````python
counts
````

````python
df['brand'].nunique()
threshold = 100
````

````python
repl= counts[counts <= threshold].index
````

## **‡¶è‡¶á ‡¶≤‡¶æ‡¶á‡¶®‡¶ü‡¶æ ‡¶¶‡¶ø‡¶Ø‡¶º‡ßá `counts` DataFrame-‡¶è‡¶∞ ‡¶∏‡ßá‡¶á index-‡¶ó‡ßÅ‡¶≤‡ßã ‡¶¨‡ßá‡¶∞ ‡¶ï‡¶∞‡¶æ ‡¶π‡¶Ø‡¶º‡ßá‡¶õ‡ßá, ‡¶Ø‡ßá‡¶ó‡ßÅ‡¶≤‡ßã‡¶∞ ‡¶Æ‡¶æ‡¶® `threshold` ‡¶•‡ßá‡¶ï‡ßá ‡¶õ‡ßã‡¶ü ‡¶¨‡¶æ ‡¶∏‡¶Æ‡¶æ‡¶®‡•§**

### **‡¶¨‡ßç‡¶∞‡ßá‡¶ï‡¶°‡¶æ‡¶â‡¶®:**
- **`counts[counts <= threshold]`** ‚Üí `counts` DataFrame ‡¶•‡ßá‡¶ï‡ßá ‡¶∂‡ßÅ‡¶ß‡ßÅ ‡¶∏‡ßá‡¶á ‡¶∞‡ßã ‡¶ó‡ßÅ‡¶≤‡ßã ‡¶®‡ßá‡¶¨‡ßá, ‡¶Ø‡ßá‡¶ó‡ßÅ‡¶≤‡ßã‡¶∞ ‡¶Æ‡¶æ‡¶® `threshold` ‡¶¨‡¶æ ‡¶§‡¶æ‡¶∞ ‡¶ö‡ßá‡¶Ø‡¶º‡ßá ‡¶ï‡¶Æ‡•§
- **`.index`** ‚Üí ‡¶è‡¶á ‡¶∞‡ßã ‡¶ó‡ßÅ‡¶≤‡ßã‡¶∞ ‡¶á‡¶®‡¶°‡ßá‡¶ï‡ßç‡¶∏ (‡¶Ö‡¶∞‡ßç‡¶•‡¶æ‡ßé ‡¶§‡¶æ‡¶¶‡ßá‡¶∞ ‡¶®‡¶æ‡¶Æ ‡¶¨‡¶æ ‡¶≤‡ßá‡¶¨‡ßá‡¶≤) ‡¶¨‡ßá‡¶∞ ‡¶ï‡¶∞‡¶¨‡ßá‡•§
- **`repl`** ‚Üí ‡¶è‡¶á ‡¶á‡¶®‡¶°‡ßá‡¶ï‡ßç‡¶∏‡¶ó‡ßÅ‡¶≤‡ßã `repl` ‡¶≠‡ßá‡¶∞‡¶ø‡¶Ø‡¶º‡ßá‡¶¨‡¶≤‡ßá ‡¶∞‡¶æ‡¶ñ‡¶æ ‡¶π‡¶≤‡ßã‡•§

````python
pd.get_dummies(df['brand'].replace(repl,'Uncommon')).astype(int)

# astype(int) use korbo karon pandas Boolean value(True,False) return kore. But amra chai jeno 0,1 e value ashe, tai astype(int) kore diyechi
````
